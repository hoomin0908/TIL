```python
import csv
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def file_open_by_numpy():
    # np.loadtxt(구분자 = ',', 데이터 타입: string)
    np_arr = np.loadtxt(r'C:\Users\SSAFY\Desktop\PJT_02_성호민\pjt\pjt02\금융\NFLX.CSV', delimiter=",", encoding='cp949', dtype=str)
    return np_arr
```

```python
arr = file_open_by_numpy()
print(arr)
```

```python
df = pd.DataFrame(arr)
df
```

```python
columns=arr[0]
arr = np.delete(arr, 0, 0)
df = pd.DataFrame(arr, columns=columns)
df
```

```python
df = df.drop(columns=['Adj Close', 'Volume'])
df
```

```python
idxs = df[df['Date'] < "2021-01-01"].index
df.drop(idxs, inplace=True)
df
```

```python
plt.figure(figsize=(9,5))
# 데이터 생성
x = df['Date']  # x 좌표값
y = df['Close'].astype(float)  # y 좌표값 #.astype(float) 실수형으로 변경

plt.plot(x, y)

# 그래프에 제목과 축 레이블 추가
plt.title('NFLX Close Price')
plt.xlabel('Date')
plt.ylabel('Close Price')


plt.xticks(rotation=45) #x축 눈금 글자 45도 회
plt.tight_layout() #그래프 안의 요소들(제목, 축 라벨, 눈금 글자)이 잘리지 않도록 여백을 자동으로 조정

# 그래프 표시
plt.show()


```

```python
df['Date'] = pd.to_datetime(df['Date'])
monthly_close = df.groupby(df['Date'].dt.to_period('M'))['Close'].mean()

plt.figure(figsize=(8,5))
x = monthly_close.index.astype(str) 
y = monthly_close.values 

plt.plot(x, y)

plt.title('Monthly Average Close Price')
plt.xlabel('Date')
plt.ylabel('Average Close Price')


plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

```python
idxs = df[df['Date'] < "2022-01-01"].index
df.drop(idxs, inplace=True)

df['Close'] = df['Close'].astype(float)
df['High'] = df['High'].astype(float)
df['Low'] = df['Low'].astype(float)

df['Date'] = pd.to_datetime(df['Date'])

plt.figure(figsize=(8,5))
x = df['Date']
y = df['Close']
y1 = df['High']
y2 = df['Low']


plt.plot(x, y, label='Close')
plt.plot(x, y1, label='High')
plt.plot(x, y2, label='Low')

plt.title('High, Low and Close Prices since January 2022')
plt.xlabel('Date')
plt.ylabel('Price')


plt.xticks(rotation=45)
plt.legend() # 범례 추가
plt.tight_layout()
plt.show()

```
